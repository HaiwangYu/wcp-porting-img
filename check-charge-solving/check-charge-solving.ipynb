```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
wcp = np.load('wcp.npy')
wct = np.load('wct.npy')
```

```python
def maplize(arr):
    m = {}
    for row in arr:
        m[tuple(row[0:8])] = row[14]
    return m

mwcp = maplize(wcp)
mwct = maplize(wct)
print (len(wcp))
print (len(wct))

diff = []
diff_rel = []
keys = []
sum_wcp = 0
sum_wct = 0
# print(wcp[0:4])
# print(wct[0:4])
for key in mwcp:
    if key in mwct:
        # print(f'{key} {mwcp[key]} {mwct[key]}')
        sum_wcp += mwcp[key]
        sum_wct += mwct[key]
        diff.append(mwcp[key]-mwct[key])
        denom = 1
        if mwcp[key] > 0:
            denom = mwcp[key]
        diff_rel.append((mwcp[key]-mwct[key])/denom)
        keys.append(key)

print(len(diff))
```

```python
sum_all_wcp = np.sum(wcp[:,14])
sum_all_wct = np.sum(wct[:,14])
print(f'{sum_all_wcp} {sum_all_wct} {np.sum(diff)/(sum_all_wcp-sum_all_wct)} {sum_wcp} {sum_wct} {np.sum(diff)} {np.sum(diff)/sum_wcp}')
fig, ax = plt.subplots()
ax.plot(diff)
plt.show()
```

```python
for i, d in enumerate(diff):
    if d > 1000 and diff_rel[i] > 0.1:
        key = keys[i]
        dr = diff_rel[i]
        print(f'{d} {dr} {key} {mwcp[key]} {mwct[key]}')
```
